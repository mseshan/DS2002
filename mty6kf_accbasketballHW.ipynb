```python
#1. READING THE CSV FILE:
import pandas as pd

acc_players_df = pd.read_csv ('/content/drive/MyDrive/Fall 24/Colab Notebooks DS 2002/acc_players-2324F edited.csv')

print("First 5 rows of the DataFrame:")
display(acc_players_df.head())
```

```python
print(acc_players_df.columns)
```

```python
#2. BASIC ANALYSIS

#calculate the total points scored by all players combined:
totalpoints = acc_players_df['PTS'].sum()
print("Total Points scored by all players:", totalpoints)

#find the player who has played the most minutes (MP):
most_minutes_player = acc_players_df.groupby('Player')['MP'].sum().idxmax()
print("Player who has played the most minutes:", most_minutes_player)

#identify the top 5 players in terms of total rebounds (TRB):
top_5_rebounds = acc_players_df.groupby('Player')['TRB'].sum().nlargest(5)
print("Top 5 players in terms of total rebounds:", top_5_rebounds)
```

```python
#3. PLAYER FILTERING

#Create a new DataFrame containing only players who played more than 500 minutes
players_above_500mins = acc_players_df[acc_players_df['MP'] > 500]
print("Players who played more than 500 minutes:")
display(players_above_500mins)


#From this filtered DataFrame, determine the player with the highest total assists (AST).
highest_assists_player = players_above_500mins.groupby('Player')['AST'].sum().idxmax()
print("Player with the highest total assists:", highest_assists_player)

#Who are the top 3 Assist Leads in the League?
top_3_assist_leads = acc_players_df.groupby('Player')['AST'].sum().nlargest(3)
print("Top 3 Assist Leads in the League:")
display(top_3_assist_leads)

#Who are the top 3 Shot Blockers?
top_3_shot_blockers = acc_players_df.groupby('Player')['BLK'].sum().nlargest(3)
print("Top 3 Shot Blockers:")
display(top_3_shot_blockers)

```

```python
#4. SCHOOL-BASED ANALYSIS

#Group the players by School and calculate the total points scored by each school.
school_points = acc_players_df.groupby('School')['PTS'].sum()
print("Total Points scored by each school:", school_points)

#Group the players by School and calculate the total assists (AST) for each team.
school_assists = acc_players_df.groupby('School')['AST'].sum()
print("Total Assists by each school:", school_assists)

#Sort the schools by total points scored and display the top 3 schools.
top_3_schools_by_points = school_points.nlargest(3)
print("Top 3 Schools by Total Points Scored:")
display(top_3_schools_by_points)

```

```python
#5. EXTRA CREDIT (BONUS)

#Create a bar chart showing the top 5 players by total points scored (PTS).
top_5_players_by_points = acc_players_df.groupby('Player')['PTS'].sum().nlargest(5)
top_5_players_by_points.plot(kind='bar', title='Top 5 Players by Total Points Scored')



```

```python
#Correlation (bonus)
import matplotlib.pyplot as plt

plt.scatter(acc_players_df['FG%'], acc_players_df['WS'])
plt.xlabel('Field Goal Percentage')
plt.ylabel('Win Shares')
plt.title('Field Goal Percentage vs. Win Shares')

correlation = acc_players_df['FG%'].corr(acc_players_df['WS'])
print("Correlation between FG% and WS:", correlation)
```
